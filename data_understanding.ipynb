```python
#Correlation matrix
corr = df.select_dtypes(include=np.number).corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matrice di correlazione")
plt.show()
```

```python
#Correlazione fra variabili
high_corr = corr.abs().unstack().sort_values(ascending=False)
high_corr[(high_corr < 1) & (high_corr > 0.9)]
```

```python
# --- Calcolo outlier_df (analisi IQR per tutte le variabili numeriche) ---

numeric_cols = df.select_dtypes(include=np.number).columns

outlier_summary = []

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    outlier_count = outliers.count()
    outlier_percentage = round((outlier_count / len(df)) * 100, 2)

    outlier_summary.append({
        "Variabile": col,
        "Outlier count": outlier_count,
        "Outlier %": outlier_percentage,
        "Q1": round(Q1, 2),
        "Q3": round(Q3, 2),
        "Lower Bound": round(lower_bound, 2),
        "Upper Bound": round(upper_bound, 2)
    })

# Crea il dataframe degli outlier
outlier_df = pd.DataFrame(outlier_summary).sort_values(by="Outlier %", ascending=False)

# Mostra le prime righe per verifica
outlier_df
```

```python
# Seleziona tutte le colonne numeriche
numeric_cols = df.select_dtypes(include='number').columns.tolist()

# Crea un boxplot per ogni variabile numerica
for col in numeric_cols:
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=df[col], color="lightblue")
    plt.title(f"Distribuzione e outlier di {col}")
    plt.show()
```
