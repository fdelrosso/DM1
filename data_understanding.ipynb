```python
# --- Calcolo outlier_df (analisi IQR per tutte le variabili numeriche) ---

numeric_cols = df.select_dtypes(include=np.number).columns

outlier_summary = []

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    outlier_count = outliers.count()
    outlier_percentage = round((outlier_count / len(df)) * 100, 2)

    outlier_summary.append({
        "Variabile": col,
        "Outlier count": outlier_count,
        "Outlier %": outlier_percentage,
        "Q1": round(Q1, 2),
        "Q3": round(Q3, 2),
        "Lower Bound": round(lower_bound, 2),
        "Upper Bound": round(upper_bound, 2)
    })

# Crea il dataframe degli outlier
outlier_df = pd.DataFrame(outlier_summary).sort_values(by="Outlier %", ascending=False)

# Mostra le prime righe per verifica
outlier_df
```
